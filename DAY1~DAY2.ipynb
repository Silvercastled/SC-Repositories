# 파이썬 구문 (Statement) 

- 특정 코드를 반복하거나, 특정 조건에 대해 원하는 코드가 실행 될 수 있도록 구성
- 조건문 : 특정 조건에 대해 코드가 실행 될 수 있도록 구성
- 반복문 : 특정 범위나 횟수에 대해 코드를 반복 수행할 수 있도록 구성

```python
 # 조건문 (if) 
data14 = 50

if data14 >= 100 : # 해당 조건을 만족하면, 바로 아래 종속된 문장이 수행
    print( '해당 데이터는 100보다 큽니다.')
    # 들여쓰기 : tab키나 space bar키를 이용해, 공백을 주어, 문장의 종속관계를 표현 
    
else: # 위의 모든 조건이 맞지 않을 경우, 아래 종속된 문장을 수행
    print('해당 데이터는 100보다 작습니다.')
```

```python
# 여러 조건을 걸어, 해당 조건에 맞는 코드를 수행할 때 
data15 = 74

if data15 >= 90 : 
    print('A')
elif data15 >= 80 :
    print('B')
elif data15 >= 70 : 
    print('C')
else: 
    print('D')
```

```python
# 데이터 필터 (Filter) / loc[] : 특정 조건(True)에 해당하는 데이터를 추출해주는 함수이다. 
# 층별화 : 범주형 데이터 항목의 특성에 따라 다른 데이터의 통계적 특성이 달라질 수 있다. 
# ex) 병원데이터에서 남성, 여성 데이터에서 생물학적으로 남.녀의 차이로 인해 근밀도나 체중에 있어서의 차이가 생길 수 있다. 

# 월랜탈비용('Amount_Month')이 10만원 이상인 고객들을 따로 뽑아 분석 
cond1 = (df1['Amount_Month'] >= 100000) # 월랜탈비용이 10만원 이상인 조건을 써줌
df1.loc[cond1] # 해당 조건에 대한 데이터를 추출 
```

```python
# 월랜탈비용('Amount_Month')이 10만원 이상이면서, 연체건수('Overdue_count')가 5회 이하인 고객명단을 추출
cond1 = (df1['Amount_Month'] >= 100000)
cond2 = (df1['Overdue_count'] <= 5)
df1.loc[cond1 & cond2]  # 두 조건을 모두 만족하는 경우 (&(And) 기호 사용)
```

```python
df1.loc[cond1 | cond2] # 두 조건 중 하나의 조건만 만족해도 데이터를 불러옴 (Or |)
```

```python
# 또 다른 동일한 방법 
cond3 = (df1['Amount_Month'] >= 100000) & (df1['Overdue_count'] <= 5)
df1.loc[cond3]
```

```python
# 또 다른 동일한 방법2 
df1.loc[(df1['Amount_Month'] >= 100000) & (df1['Overdue_count'] <= 5)]
```

```python
# 판매유형(Sales_Type)이 '렌탈'에 해당하는 데이터만 추출 
cond4 = (df1['Sales_Type'] == '렌탈')
df1.loc[cond4]
```

```python
# 고객들의 유입경로 (Channel)이 '영업방판', '홈쇼핑/방송', '대형마트A' 여러 항목에 대해 데이터를 추출 

cond5 = (df1['Channel'] == '영업방판')
cond6 = (df1['Channel'] == '홈쇼핑/방송')
cond7 = (df1['Channel'] == '대형마트A')
df1.loc[cond5|cond6|cond7]
```

```python
# isin() : 특정 리스트 내의 데이터가 있다면 True, 없다면 False 
cond8 = (df1['Channel'].isin( ['영업방판', '홈쇼핑/방송', '대형마트A']))
df1.loc[cond8]
```

```python
# 연습 문제 : 결제은행('Bank')이 '현대카드'에 해당하는 계약 명단 중, 월랜탈비용이('Amount_Month') 150,000이상인 상위 50명의 계약명단을 
# 추출하여 result3.xlsx에 저장
cond9 = (df1['Bank'] == '현대카드')
cond10 = (df1['Amount_Month'] >= 150000)
df1.loc[cond9 & cond10].head(50).to_excel('result3.xlsx')


```
